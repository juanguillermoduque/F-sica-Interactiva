**SONIDO**


El sonido es una onda mecánica que se propaga a través de un medio material (como aire, agua o sólidos) debido a la vibración de las partículas del medio. Es una onda longitudinal que consiste en oscilaciones sucesivas de compresión y rarefacción del medio por el cual viaja. Para que el sonido se genere y propague, es necesario un medio elástico, ya que el sonido no se puede transmitir en el vacío. Las características perceptibles del sonido para los seres humanos incluyen el tono, que está determinado por la frecuencia de la onda sonora, y el volumen, que depende de la amplitud de la onda.

**Nivel de intensidad sonora:**
El nivel de intensidad sonora, también conocido como nivel de presión sonora, es una medida cuantitativa del sonido percibido, expresada en decibelios (dB). Este nivel es una medida logarítmica de la ratio entre la intensidad de un sonido y un valor de referencia, lo cual significa que describe la intensidad de un sonido en una escala que imita la respuesta del oído humano al sonido. Aquí está cómo se define y calcula:

### Naturaleza del Sonido
- **Onda Mecánica**: El sonido es una onda mecánica longitudinal que necesita un medio para propagarse.
- **Frecuencia**: La frecuencia del sonido determina su tono. Se mide en Hertz (Hz). Los sonidos audibles para el oído humano están en el rango de aproximadamente 20 Hz a 20,000 Hz.
- **Amplitud**: La amplitud de la onda de sonido está relacionada con su intensidad o volumen. Se mide en decibelios (dB).

### Propagación del Sonido
- **Velocidad**: La velocidad del sonido depende del medio a través del cual se propaga. Por ejemplo, en el aire a 20 °C, la velocidad es aproximadamente 343 m/s. En el agua, es alrededor de 1,480 m/s.
- **Reflexión**: Cuando una onda sonora encuentra un obstáculo, puede reflejarse, produciendo ecos.
- **Refracción**: El sonido puede cambiar de dirección al pasar de un medio a otro, debido a la variación en la velocidad de propagación.
- **Difracción**: Las ondas sonoras pueden doblarse alrededor de obstáculos o pasar a través de aberturas, permitiendo que el sonido se escuche incluso si la fuente no es directamente visible.

### Características del Sonido
- **Timbre**: La calidad o el color del sonido, que permite distinguir entre diferentes fuentes sonoras que tienen la misma frecuencia e intensidad.
- **Tono**: La percepción subjetiva de la frecuencia del sonido. Un tono alto tiene una frecuencia mayor, y un tono bajo tiene una frecuencia menor.
- **Intensidad**: La potencia del sonido por unidad de área, que se percibe como el volumen del sonido.

### Aplicaciones del Sonido
- **Comunicación**: El sonido es fundamental para la comunicación humana a través del habla.
- **Música**: La manipulación del sonido en términos de tono, timbre e intensidad para crear música.
- **Tecnología**: Aplicaciones en sonar, ultrasonido médico, acústica arquitectónica, y más.

### Fenómenos Acústicos
- **Efecto Doppler**: El cambio en la frecuencia del sonido debido al movimiento relativo de la fuente y el receptor.
- **Resonancia**: La amplificación de la amplitud de las ondas sonoras cuando la frecuencia de la onda coincide con la frecuencia natural del sistema.

**OIDO HUMANO**

*Funcionamiento del oído humano:*
El oído humano es un órgano complejo y fascinante que nos permite percibir el mundo que nos rodea a través del sonido. Se divide en tres partes principales:

1. Oído externo:

Compuesto por el pabellón auricular, el conducto auditivo externo y el tímpano.
El pabellón auricular recoge las ondas sonoras y las dirige hacia el conducto auditivo externo.
El conducto auditivo externo termina en el tímpano, una membrana delgada y vibrante.
2. Oído medio:

Contiene tres huesecillos diminutos: el martillo, el yunque y el estribo.
El tímpano transmite las vibraciones a los huesecillos, amplificando las ondas sonoras.
Los huesecillos conectan con la ventana oval, una abertura en la pared ósea del oído interno.
3. Oído interno:

Contiene la cóclea, una estructura en forma de caracol llena de líquido.
La ventana oval transmite las vibraciones al líquido de la cóclea, creando ondas.
Las células ciliadas, ubicadas en la membrana basilar de la cóclea, se convierten en impulsos nerviosos.
El nervio auditivo transporta estos impulsos nerviosos al cerebro, donde se interpretan como sonido.

*Etapas del proceso auditivo:*

Recepción de ondas sonoras: Las ondas sonoras ingresan al oído externo a través del pabellón auricular y viajan por el conducto auditivo externo.
Transmisión de vibraciones: Las ondas sonoras hacen vibrar el tímpano, que a su vez transmite las vibraciones a los huesecillos del oído medio.
Amplificación de las ondas: Los huesecillos amplifican las vibraciones y las transmiten a la ventana oval.
Creación de ondas en el líquido: La ventana oval transmite las vibraciones al líquido de la cóclea, creando ondas.
Transformación en impulsos nerviosos: Las células ciliadas detectan las ondas y se convierten en impulsos nerviosos.
Transmisión al cerebro: El nervio auditivo transporta los impulsos nerviosos al cerebro.
Interpretación como sonido: El cerebro interpreta los impulsos nerviosos como sonido, permitiéndonos percibir los diferentes sonidos del entorno.
