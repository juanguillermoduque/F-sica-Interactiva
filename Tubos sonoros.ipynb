**TUBOS SONOROS**

Los tubos sonoros son columnas de aire confinadas dentro de un tubo que pueden vibrar para producir ondas sonoras. Estos tubos pueden ser abiertos en ambos extremos o cerrados en un extremo. La teoría de los tubos sonoros se basa en los principios de resonancia y ondas estacionarias en columnas de aire.

1. **Ondas Estacionarias**

* En un tubo sonoro, el aire puede vibrar en una serie de modos, cada uno correspondiente a una frecuencia natural del tubo.

* Las ondas estacionarias se forman debido a la interferencia constructiva entre las ondas que viajan en direcciones opuestas dentro del tubo.

2. **Modos de Resonancia**

* Los modos de resonancia en un tubo abierto en ambos extremos tienen nodos en los extremos y antinodos en el centro.

* En un tubo cerrado en un extremo y abierto en el otro, hay un nodo en el extremo cerrado y un antinodo en el extremo abierto.

3. **Frecuencias de Resonancia**

* La frecuencia fundamental y los armónicos de un tubo dependen de su longitud y de si el tubo es abierto o cerrado.

* Para un tubo abierto, las frecuencias de resonancia son múltiplos enteros de la frecuencia fundamental $ f = n \cdot v /(2 \cdot L)$ donde $n$ es un entero positivo,
$v$ es la velocidad del sonido en el aire, y
$L$  es la longitud del tubo.

* Para un tubo cerrado en un extremo, las frecuencias son múltiplos impares de la frecuencia fundamental $ f = (2 \cdot n +1) \cdot v /(4 \cdot L)$ donde $n=1,2,3,\cdots $.

4. **Efecto del Medio**

* La velocidad del sonido y, por lo tanto, las frecuencias de resonancia pueden variar dependiendo del medio (aire, helio, etc.) dentro del tubo.

**TUBO ABIERTO**
Son tubos que tienen un extremo abierto y otro cerrado. El extremo abierto permite que el aire se mueva libremente dentro y fuera del tubo. Estos tubos resonarán a ciertas frecuencias que dependen de la longitud del tubo y de la velocidad del sonido en el medio (generalmente el aire).

**TUBO CERRADO**
