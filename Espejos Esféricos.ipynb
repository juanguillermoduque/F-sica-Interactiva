#**Espejos Esféricos**
Los espejos esféricos son superficies reflectantes que tienen la forma de una parte de una esfera. Existen dos tipos principales de espejos esféricos:

###**Espejos cóncavos:** La superficie reflectante está en el interior de la esfera.

###**Espejos convexos:** La superficie reflectante está en el exterior de la esfera.

#**Espejo Convexo**
Un espejo convexo es un tipo de espejo esférico cuya superficie reflectante está en el exterior de una esfera. Estos espejos son conocidos por divergir la luz que incide sobre ellos, lo que significa que los rayos de luz que se reflejan parecen provenir de un punto focal virtual detrás del espejo.

image.png

##**Características de los Espejos Convexos**
**Divergencia de la luz:** Los rayos de luz que inciden paralelamente al eje principal se reflejan y divergen como si vinieran de un punto focal virtual detrás del espejo.

**Imagen virtual:** Las imágenes formadas por espejos convexos son siempre virtuales, derechas y más pequeñas que el objeto real.

**Amplitud de visión:** Los espejos convexos proporcionan un campo de visión más amplio, por lo que son utilizados comúnmente en retrovisores de automóviles y en tiendas para vigilancia.



```python
# @title Simulación de un espejo convexo (trazado de rayos)
import matplotlib.pyplot as plt
import numpy as np

# Configuración de la figura
fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')

# Definición del espejo convexo y los rayos
def draw_convex_mirror():
    theta = np.linspace(0, np.pi, 100)
    x = np.cos(theta)
    y = np.sin(theta)
    ax.plot(x, y, 'k', label='Espejo Convexo')

def draw_rays():
    # Rayos incidentes
    ax.plot([2, 0.5], [1, 1], 'r', label='Rayo 1')
    ax.plot([2, -0.5], [0.5, 0.5], 'b', label='Rayo 2')

    # Rayos reflejados
    ax.plot([0.5, -1], [1, 2], 'r--')
    ax.plot([-0.5, -2], [0.5, 1], 'b--')

    # Puntos de origen y foco virtual
    ax.plot([0], [0], 'ko', label='Centro del Espejo')
    ax.plot([-1], [0], 'ko', label='Foco Virtual')

# Dibujar el espejo y los rayos
draw_convex_mirror()
draw_rays()

# Configuración de la gráfica
ax.legend()
ax.grid(True)
ax.set_xlim([-2, 2])
ax.set_ylim([0, 2])
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
plt.title('Trazado de Rayos en un Espejo Convexo')
plt.show()
```

```python
# @title Cálculo de la imagen en un espejo convexo
import matplotlib.pyplot as plt
import numpy as np

# Función para dibujar el espejo convexo
def draw_convex_mirror():
    theta = np.linspace(0, np.pi, 100)
    x = np.cos(theta)
    y = np.sin(theta)
    plt.plot(x, y, 'k', label='Espejo Convexo')
    plt.plot(-x, y, 'k--')

# Función para trazar los rayos y la imagen
def draw_rays_and_image(f, d_o):
    # Cálculo de la distancia de la imagen
    d_i = 1 / ((1 / f) - (1 / d_o))
    M = -d_i / d_o

    # Dibujar el objeto
    plt.plot([d_o, d_o], [0, 1], 'g', label='Objeto')
    plt.plot([d_o], [1], 'go')

    # Rayos incidentes y reflejados
    plt.plot([d_o, 0], [1, 1], 'r', label='Rayo 1 (Incidente)')
    plt.plot([d_o, 0], [1, 0.5], 'b', label='Rayo 2 (Incidente)')
    plt.plot([0, d_i], [1, M], 'r--', label='Rayo 1 (Reflejado)')
    plt.plot([0, d_i], [0.5, M], 'b--', label='Rayo 2 (Reflejado)')

    # Dibujar la imagen virtual
    plt.plot([d_i, d_i], [0, M], 'm', label='Imagen Virtual')
    plt.plot([d_i], [M], 'mo')

# Parámetros del espejo y del objeto
f = -10  # Distancia focal (negativa para espejo convexo)
d_o = 20  # Distancia del objeto al espejo

# Configuración de la gráfica
plt.figure(figsize=(10, 5))
draw_convex_mirror()
draw_rays_and_image(f, d_o)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Simulación de un Espejo Convexo')
plt.xlabel('Eje X (Distancia)')
plt.ylabel('Eje Y (Altura)')
plt.xlim([-15, 25])
plt.ylim([-5, 5])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

#**Espejos Cóncavos**
Un espejo cóncavo es un tipo de espejo esférico cuya superficie reflectante está en el interior de una esfera. Estos espejos son conocidos por converger la luz que incide sobre ellos, lo que significa que los rayos de luz que se reflejan tienden a juntarse en un punto focal real delante del espejo.

##**Características de los Espejos Cóncavos**
**Convergencia de la luz:**
Los rayos de luz que inciden paralelamente al eje principal se reflejan y convergen en un punto focal real delante del espejo.

**Imágenes reales y virtuales:** Dependiendo de la posición del objeto con respecto al espejo, se pueden formar imágenes reales (invertidas y de diferentes tamaños) o imágenes virtuales (derechas y más grandes que el objeto real).

**Amplitud de visión:** Los espejos cóncavos proporcionan un campo de visión más estrecho y se utilizan comúnmente en telescopios, reflectores y espejos de maquillaje.

```python
# @title Simulación de un Espejo Cóncavo (Trazado de Rayos)
import matplotlib.pyplot as plt
import numpy as np

# Función para dibujar el espejo cóncavo
def draw_concave_mirror():
    theta = np.linspace(0, np.pi, 100)
    x = -np.cos(theta)
    y = np.sin(theta)
    plt.plot(x, y, 'k', label='Espejo Cóncavo')
    plt.plot(-x, y, 'k--')

# Función para trazar los rayos y la imagen
def draw_rays_and_image(f, d_o):
    # Cálculo de la distancia de la imagen
    d_i = 1 / ((1 / f) - (1 / d_o))
    M = -d_i / d_o

    # Dibujar el objeto
    plt.plot([-d_o, -d_o], [0, 1], 'g', label='Objeto')
    plt.plot([-d_o], [1], 'go')

    # Rayos incidentes y reflejados
    plt.plot([-d_o, 0], [1, 1], 'r', label='Rayo 1 (Incidente)')
    plt.plot([-d_o, 0], [1, 0], 'b', label='Rayo 2 (Incidente)')
    plt.plot([0, d_i], [1, 0], 'r--', label='Rayo 1 (Reflejado)')
    plt.plot([0, d_i], [0, 0], 'b--', label='Rayo 2 (Reflejado)')

    # Dibujar la imagen real
    plt.plot([d_i, d_i], [0, M], 'm', label='Imagen Real')
    plt.plot([d_i], [M], 'mo')

# Parámetros del espejo y del objeto
f = 10  # Distancia focal (positiva para espejo cóncavo)
d_o = 20  # Distancia del objeto al espejo

# Configuración de la gráfica
plt.figure(figsize=(10, 5))
draw_concave_mirror()
draw_rays_and_image(f, d_o)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Simulación de un Espejo Cóncavo')
plt.xlabel('Eje X (Distancia)')
plt.ylabel('Eje Y (Altura)')
plt.xlim([-25, 15])
plt.ylim([-5, 5])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
